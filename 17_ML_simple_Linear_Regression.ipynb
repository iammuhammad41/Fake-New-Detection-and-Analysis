```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
# Step 1
# dataloader
dataset = pd.read_csv('/content/drive/MyDrive/fake_news_detection/data/Salary_Data.csv')
# print(dataset)
print(dataset.head())

# Step 2
# split dataset into training data and testing data

split_data1 = dataset[['YearsExperience']]
split_data2 = dataset[['Salary']]

from sklearn.model_selection import train_test_split
# X_test, X_train, y_test, y_train= train_test_split(split_data1, split_data2, test_size=1/5, random_state=0)
X_train, X_test, y_train, y_test = train_test_split(split_data1, split_data2, test_size=1/5, random_state=0)

#fit Linear Regression Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model = model.fit(X_train, y_train)
#
# # we can also use for above two lines
# model = LinearRegression().fit(X_train, y_train)
print(model)
# print('1')

# plotting
''''----'''
#
# m = stats.linregress(split_data1, split_data2)
# t = [m.slope * i + m.intercept for i in X_train]
# dataset.insert(len(dataset.columns),'Salary',t)
# ax = dataset.plot.scatter(dataset[['YearsExperience']],
#                        dataset[['Salary']])
# dataset.plot.line(dataset[['YearsExperience']],
#                dataset[['Salary']],
#                color='red',
#                ax=ax,
#                figsize = (16,8))

'''----'''
plt.scatter(X_train, y_train)
# plt.scatter(X_test, y_test)
# print('2')
# predicting the data
# plt.plot(X_train, y_train)
plt.plot(X_train, model.predict(X_train), color='Green')
# print('2.1')
plt.title('Train Plot')
plt.xlabel('Experiment')
# print('2.2')
plt.ylabel('Dadqah')
# print('2.3')

# print('3')
plt.show()

# print('4')
plt.scatter(X_test, y_test)

# predicting the data
plt.plot(X_test, model.predict(X_test), color='Green')
plt.xlabel('Experiment')
plt.ylabel('Dadqah')
plt.title('Test Plot')
print('5')
plt.show()

# testing and evaluating the model
eval_train = model.score(X_train, y_train)
eval_test = model.score(X_test, y_test)
# print('6')
print('Score for Training Data', eval_train)
print('Score for Testing Data', eval_train)

# prediction of unknown values
pred = model.predict([[5], [10], [20], [30]])
# pred = model.predict([[10]])
# pred = model.predict(X_test)
# print('7')
print(pred)

x = ([[5], [10], [20], [30]])
# x = ([[1], [2], [3], [4], [5]])
# print('8')
print(model.predict(x))
```
